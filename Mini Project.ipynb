```python
import csv
import matplotlib.pyplot as plt
with open('modifiedsheet.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    d = [j['D.O. (mg/l)'] for j in reader]
    d1 = d[:150]
    d1 = [float(i) for i in d1]

with open('modifiedsheet.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    t = [i['Temp'] for i in reader]
    t1 = t[:150]
    t1 = [float(i) for i in t1]

with open('modifiedsheet.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    c = [i['CONDUCTIVITY (µmhos/cm)'] for i in reader]
    c1 = c[:150]
    c1 = [float(i) for i in c1]

with open('modifiedsheet.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    p = [j['PH'] for j in reader]
    p1 = p[:150]
    p1 = [float(i) for i in p1]

with open('modifiedsheet.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    bo = [j['B.O.D. (mg/l)'] for j in reader]
    bo1 = bo[:150]
    bo1 = [float(i) for i in bo1]

x1 = [i * i for i in t1]
x2 = [t1[i] * d1[i] for i in range(len(t1))]
m = ((len(t1) * sum(x2)) - sum(t1) * sum(d1)) / ((len(t1) * sum(x1)) - sum(t1) ** 2)
b = (sum(d1) - (m * sum(t1))) / len(t1)

x = [i for i in range(1, 151)]

dd = [round((m*i + b), 2) for i in t1]

plt.plot(x, d1, x, dd)
plt.xlabel("Temp")
plt.ylabel("DO")
plt.show()

y1 = [i * i for i in t1]
y2 = [t1[i] * p1[i] for i in range(len(t1))]
m1 = ((len(t1) * sum(y2)) - sum(t1) * sum(p1)) / ((len(t1) * sum(y1)) - sum(t1) ** 2)
b1 = (sum(p1) - (m1 * sum(t1))) / len(t1)

dd1 = [round((m1*i + b1), 2) for i in t1]

plt.plot(x, p1, x, dd1, )
plt.xlabel("Temp")
plt.ylabel("PH")
plt.show()

z1 = [i * i for i in t1]
z2 = [t1[i] * c1[i] for i in range(len(t1))]
m3 = ((len(t1) * sum(z2)) - sum(t1) * sum(c1)) / ((len(t1) * sum(z1)) - sum(t1) ** 2)
b3 = (sum(c1) - (m * sum(t1))) / len(t1)

dd2 = [round((m3*i + b3), 2) for i in t1]

plt.plot(x, c1, x, dd2, )
plt.xlabel("Temp")
plt.ylabel("Conductivity")
plt.show()

w1 = [i * i for i in t1]
w2 = [t1[i] * bo1[i] for i in range(len(t1))]
m4 = ((len(t1) * sum(w2)) - sum(t1) * sum(bo1)) / ((len(t1) * sum(w1)) - sum(t1) ** 2)
b4 = (sum(bo1) - (m4 * sum(t1))) / len(t1)

x = [i for i in range(1, 151)]

dd3 = [round((m4*i + b4), 2) for i in t1]

plt.plot(x, bo1, x, dd3, )
plt.xlabel("Temp")
plt.ylabel("BOD")
plt.show()
```
